```python
import os
def extract_file_numbers(directory):
    extracted_numbers = {}
    for filename in os.listdir(directory):
        # print('filename', filename)
        file_name = filename.split("/")[0]
        number_str = file_name.split(".")[0]

        image_path = directory + "/" + file_name
            
        extracted_numbers[int(number_str)]= Image.open(image_path)

    return extracted_numbers

# x = extract_file_numbers("/tmp/audio_0")

dir_list = ["audio_0", "audio_1", "audio_2", "audio_3", "audio_4", "audio_5", "audio_6", "audio_61", "audio_62", "audio_63", "audio_64", "audio_65", 
"audio_66", "audio_67", "audio_68", "audio_69", "audio_7"]

full_dict = {}
for dir_ in dir_list:
    dir_full = "/tmp/" + dir_
    print(dir_full)

    full_dict[dir_] = extract_file_numbers(dir_full)
full_dict
```
